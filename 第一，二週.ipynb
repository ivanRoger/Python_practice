### 拍拍機器人v1.0

```python
message = ""
while message != "bye":
    message = input(">>")
    print("拍拍")
```

###  字串簡簡單單格式化 


```python
name = "Ivan"
place = "台中"
```

```python
message = "你好，來自" + place + "的" + name + "！"
```

```python
print(message)
```

```python
message = "你好，來自{}的{}！".format(place, name)
```

```python
print(message)
```
